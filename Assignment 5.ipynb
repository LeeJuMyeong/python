## 설명

> 다형성(Polymorphism)이란 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 하는 기능입니다.


```python
#-*- coding: utf-8 -*-
class Figure:
    def sideCount(self):
        return None
class Triangle(Figure):
    def sideCount(self):
        return 3
class Quadrangle(Figure):
    def sideCount(self):
        return 4
class Pentagon(Figure):
    def sideCount(self):
        return 5

for each in (Figure(), Triangle(), Quadrangle(), Pentagon()):
    print each.sideCount()
```

```python
class Counter:
    def __init__(self, count, step=1):
        self.count = count
        self.step = step
    def __str__(self):
        return "[Count (step: %d)] %d" % (self.step, self.count)
    def incr(self):
        self.count += self.step
    def __call__(self):
        self.count += self.step
    def __add__(self, num):
        self.count += num
        return self
    def __sub__(self, num):
        self.count -= num
        return self
    
    def __cmp__(self, other):
        return self.count - other 
    
c = Counter(10)
d = Counter(10, 2)

print c
print d
print 

c.incr()
d.incr()

print c
print d
print 

c()
d()

print c
print d
print 

c = c + 5
d = d - 5

print c
print d
print 

print c > 10
print d > 10
print c < 10
print d < 10
print c == 17
print d != 9
print 
```

_[4번 문제]_  

다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 \_\_init\_\_(), \_\_str()\_\_(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.  

- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

 > Set 자료형을 사용할 수 있는 클래스를 만들기 위해서는 정의하려는 클래스(MySet)에 list 클래스를 종속해야한다. 이렇게 만들어진 MySet 클래스에서 \_\_init\_\_()함수는 생성자를 만드는 함수로써, 지금은 Set 자료형을 정의할 수 있게 정의해야합니다. 그러므로, 생성자에 넣을 수 있는 인자는 자료형 list 받게 합니다. 그리고 그 리스트(l)의 내용을 인스턴스 객체에 넣습니다. 단, 이 경우 l에 중복 값이 있으면 중복값을 제거해줄 수 있어야하기 때문에, eliminate\_duplicate() 함수를 정의합니다.  

> 이 함수는 인자를 리스트로 받고, 받은 리스트를 반복문과 조건문을 사용해서 중복된 값을 제거하도록 합니다. 하지만 이 함수는 MySet 클래스에서 인스턴스 객체가 없어도 사용될 수 있게 만들어야하므로, @staticmethod라고 지정해야합니다.  

> 그리고 마지막으로, \_\_str\_\_()함수를 사용하여 객체에 있는 값들을 반환할 수 있게 문자열로 리스트 값들을 문자열로 나열하고 반환하도록 정의합니다.

_[5번 문제]_  
4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오  

- 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.  
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.  

> \>\>\> u = s | t  
\>\>\> print u  
MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}  

- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.  

> \>\>\> u = s & t   
\>\>\> print u   
MySet: {2, 3}   

- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.  

> \>\>\> s = MySet([1, 2, 3])   
\>\>\> t = MySet([3, 4, 5])  
\>\>\> u = s - t  
\>\>\> print u  
MySet: {1, 2}   


```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    
    
    # 요구사항 1. 답
    def __or__(self, other):    # |으로 두 집합의 합집합을 반환하기 위해서는 __or__()함수를 정의해야함
        l = []                  # 합집합이기때문에 빈 리스트 하나에 self와 other에 있는 값들을 리스트에 sppend하면 됨
        for e in self:
            l.append(e)
        for e in other:
            l.append(e)
            
        result = MySet(l)        # 마지막으로 result 라는 MySet() 객체에 리스트를 포함하여 반환하도록 함
        return result         

    # 요구사항 2. 답
    def __and__(self, other):    # &으로 두 집합의 교집합을 반환하기 위해서는 __and__()함수를 정의해야함
        l = []                   # 교집합은 두 객체가 가지고 잇는 값들 중 동일하게 갖고 잇는 값들만을 리스트로 만들어야함
        for e in other:          # 그러므로 일단 other에 있는 값들 중 
            if e in self:        # self에 있는 값들과 같은 것을 리스트에 추가하도록 함
                l.append(e)
        result = MySet(l)
        return result
    
    
    # 요구사항 3. 답
    def __sub__(self, other):    # -으로 두 집합의 차집합을 반환하기 위해서는 __sub__()함수를 정의해야함
        l = []                   # 차집합은 위의 교집합과 반대로 조건문을 사용하면 됨 
        for e in self:           # 단, 주의해야할 점은 self에 있는 값들을 반복문으로 비교하고
            if e not in other:   # other에 있는 값들 중에 없는 값들만 리스트에 추가하게 하는 방법을 사용해야함
                l.append(e)
        result = MySet(l)
        return result
    
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
    print 
    
u = s | t
print u
print


u = s & t
print u
print 


s = MySet([1, 2, 3])
t = MySet([3, 4, 5])
u = s - t
print u
```

프로젝트 오일러 문제 16  

215 = 32768 의 각 자리수를 더하면 3 + 2 + 7 + 6 + 8 = 26 입니다.  

21000의 각 자리수를 모두 더하면 얼마입니까?  


```python
def cipher_sum(num):
    str_num = str(num)
    result = 0
    for i in range(0, len(str_num)):
        result += int(str_num[i])
    return result
    
print cipher_sum(21000)
```
