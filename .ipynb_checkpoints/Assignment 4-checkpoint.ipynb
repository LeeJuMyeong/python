_[3번 문제]_  

range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.  

- 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.  
    > range(stop)  
    range(start, stop[, step])  
- 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.  
- 3) frange 사용 예  
    - 3-1) frange(0.5)  
        - [0.0, 0.1, 0.2, 0.3, 0.4]  
    - 3-2) frange(1.0, 2.0)  
        - [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]  
    - 3-3) frange(2.2, 4.0, 0.5)  
        - [2.2, 2.7, 3.2, 3.7]  

```python
def frange(start = 0.0, stop = None, step = 0.1):
    a = start
    b = stop
    c = step
    l = []
    
    if stop is None:
        a, b = b, a
    if a is None:
        a = 0.0
    while a < b:
        l.append(round(a, 5))
        a += c
    print l
    # return l
        
frange(0.5)
frange(1.0, 2.0)
frange(2.2, 4.0, 0.5)
```

## 설명

> 이 문제는 default 값을 지정하여 range와 유사한 사용자가 frange라는 함수를 만드는 것입니다. 일단 저는 start와 step에는 각 각 default 값을 문제에서 제시한 숫자로 지정했고, 남은 stop의 default 값을 None으로 설정했습니다. 이유는 start와 step만 디폴드 값을 지정하게 되면 오류가 발생하기 때문입니다. 그리고 range 함수는 인자를 하나만 받을 경우 그 인자 값이 stop으로 되야하기 때문에 첫번째 if 문에서 stop 값이 디폴트 값으로 나올 경우, 그 start, stop의 값을 저장하는 a, b를 swap 하게 됩니다. 그리고 a(start 값)이 None일 경우, 기본값인 0.0으로 바꿔주는 조건문을 두번째에 넣었습니다. 그리고 마지막으로, start,stop,step 값을 저장한 a,b,c를 이용해서 range와 같이 리스트 형태로 값을 출력하게 됩니다. 이때 float 값이 부동소수점 연산시 이상한 점이 있어서 round 함수를 써서 값을 다시 잡아줍니다.

_[4번 문제]_  

가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오  

- sum() 사용 예  
    - sum()  
        - 0  
    - sum(1, 2)  
        - 3  
    - sum(1, 2, 3, 4, 5)  
        - 15  
    - sum(1, 5, 7, 2, -10)  
        - 5  


```python
def sum(*args):
    num = 0
    for i in args:
        num += i
    print num
    
sum()
sum(1,2)
sum(1,2,3,4,5)
sum(1,5,7,2,-10)
```

---

> 이 문제는 가변인수를 사용해서 함수를 정의했습니다. 일단 가변인수인 * args를 sum의 인자로 넣고, 반복문을 사용해서 args에 들어있는 숫자를 모두 num에 더합니다. 그리고 num를 출력하면 됩니다.

_[5번 문제]_  
여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.  

- 1) 다음에 제시되는 함수들을 모두 이용해야 한다.  
    - split
    - map
    - join
- 2) myinitial() 함수 사용 예
    - myinitial("as soon as possible")
        - ASAP

```python
def myinitial(s):
    list_str = s.split()
    list_initial = []
    for i in list_str:
        list_initial.append(i[0])
        
    list_initial = map(lambda x: chr(ord(x) - 32), list_initial)
    
    join_list = "".join(list_initial)
    
    print join_list
    
myinitial("as soon as possible")
    
```

## 설명

> 이 문제는 split, map, join을 모두 써서 함수를 만들었습니다. 일단 각 단어를 나누기 위해서 split를 사용하여 띄어쓰기를 기준으로 단어로 분류하였고, 나누어진 값들에서 첫글자만 가져오기 위해서 for문을 사용하였습니다. 그리고 lambda와 map 함수를 사용하여 저장된 첫 글자들을 모두 대문자로 바꾸었고, 그 값들을 join을 통해서 합칩니다. 그리고 그 값을 출력였습니다.

음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.  

- [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
def myfact(n):
    fact_num = 1
    if n == 1:
        return n
    if n > 1:
        fact_num = n * myfact(n-1)
        return fact_num

print myfact(976)
```

## 설명   

> 이 문제에서 함수를 재귀적으로 정의하였습니다. 일단 팩토리얼 값을 얻는 것이기 때문에 n(인자값)이 1일 경우에는 바로 n을 반환하고 1보다 큰 값일 경우에는 fact_num에 n * myfact(n-1)을 하는 데, 이렇게 할 경우 함수를 재귀적으로 호출하게 됩니다. 옵션으로 기다릴수 있을 만큼의 수행시간을 직접 기다려보면서 늘리라고 하셨는데, 일단 수행시간이 오래 걸리진 않지만, maximum에 재귀에 걸려서 오류가 나서 그 전까지의 수만 출력해봤습니다.

프로젝트 오일러 문제 10  

10 이하의 소수를 모두 더하면 2 + 3 + 5 + 7 = 17 이 됩니다.
  
이백만(2,000,000) 이하 소수의 합은 얼마입니까?  

```python
def sum_prime(n):
    l = []
    if(n <= 1):
        return 0;
    for i in range(1, n):
        l.append(None)
    for i in range(1, (n+1)/2): 
        l[i] = True
    sum_n = 2
    for i in range(1, (n+1)/2):
        
        if(l[i] == True):
            p = 2*i+1  
            sum_n += p
            
            j = 3
            while j*p <= n:
                l[(j*p-1)/2] = False
                j += 2
    return sum_n

print sum_prime(2000000)
```

## 설명  

> 이 문제에서는 무지 큰 범위의 숫자 이하 소수의 합을 얻는 것이 중요했습니다. 그 전에 소수를 찾는 방법 중 "에라토스테네스의 체"라는 공식을 이용해보았습니다. 에라토스테네스의 체는 max 값을 기준으로 max 값보다 작은 소수의 값의 제곱 중 제일 큰 값(p_max)을 기준으로, p_max 값보다 작은 수의 소수 값들의 배수를 지우고 남는 수가 모두 소수라고 판단하는 것입니다.
  
> 일단 함수를 정의하는 데 인자로 범위를 지정할 수(n)를 입력합니다. 일단 n 값이 1과 같거나 작을 경우에는 0을 반환합니다. 그리고 for문을 사용하여 n의 크기 만큼의 None 값이 들어있는 리스트 l을 생성합니다. 그리고 다시 리스트 l에 현재 크기의 반만큼의 값들을 True의 값으로 만듭니다. (여기서 현재 크기의 반만큼 하는 이유는 2의 배수, 즉 짝수는 소수가 될 수 없기 때문에(2는 제외) 전체적인 속도를 고려해서 추가한 것입니다.) 다음으로 sum_n의 값을 2로 저장합니다.(2의 배수중 2만 소수이기때문에)그리고 아까와 마찬가지로 리스트 l에 현재 크기의 반만큼의 값들을 반복문으로 돌리고, 그 중 조건문으로 True인 값들일 경우에만 그 값의 원래 값(p)을 sum_n에 저장합니다. 마지막으로 다음은 반복문 내부에 다시 반복문을 넣어서 소수들의 배수들의 값(소수들의 배수를 찾을 때도, 짝수 즉, 2를 제외한 2의 배수는 어차피 안들어있기때문에 무시하게 반복)들에 대해서 False 값으로 바꾸게 한 후, sum_n을 출력합니다.

프로젝트 오일러 문제 12  

1부터 n까지의 자연수를 차례로 더하여 구해진 값을 삼각수라고 합니다.  
예를 들어 7번째 삼각수는 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28이 됩니다.  
이런 식으로 삼각수를 구해 나가면 다음과 같습니다.  
  
    1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...  
이 삼각수들의 약수를 구해봅시다.  

     1: 1  
     3: 1, 3  
     6: 1, 2, 3, 6  
    10: 1, 2, 5, 10  
    15: 1, 3, 5, 15  
    21: 1, 3, 7, 21  
    28: 1, 2, 4, 7, 14, 28  
위에서 보듯이, 5개 이상의 약수를 갖는 첫번째 삼각수는 28입니다.  
  
그러면 500개 이상의 약수를 갖는 가장 작은 삼각수는 얼마입니까?  


```python
def div_num(n):
    temp = 0
    num_cal = 1
    num = n
    
    while(num % 2 == 0):
        num = num/2;
        temp += 1
        
    num_cal *= temp+1
    
    i = 3
    while i*i < num: 
        temp = 0
        while (num % i == 0):
            num = num / i
            temp += 1
        num_cal *= temp+1
        
        if(n == 1):
            break
        i += 2
        
    if(n != 1):
        num_cal *= 2
    
    return num_cal 
    
d_n = 0

num = 0
sum_n = 0 

while 1:
    num += 1
    sum_n += num 
    if div_num(sum_n) >= 500:
        d_n = sum_n
        break
print d_n
```

## 설명

> 이 문제는 소인수 분해의 갯수를 구하는 것이 포인트인 문제였습니다. 일단 무한루프를 통해서 삼각수를 구하게 만들고 이 값들을 div_num라는 함수에 넣습니다. div_num는 인자로 받은 값의 약수의 갯수를 구하는 함수입니다. 이때, 약수의 개수를 구하는 방법은 일단 인자(n)값을 num에 넣고 num값이 2로 나누어지는지 확인한 후, 나누어지면 temp를 올리고, 나누어지지 않을 때까지 반복합니다. 처음 2만 따로 반복을 해주고, 다음 3부터는 step을 2씩올리면서 num값이 1이 될때까지 혹은 i * i < num 아닐때까지 반복하고, 그 후 나온 num_cal 값을 반환합니다. 그리고 그 값이 500이상인지 확인후, 아니면 다시 반복하고, 맞다면 break를 해서 반복문에 나온후 d_n을 출력합니다.

프로젝트 오일러 문제 14  

양의 정수 n에 대하여, 다음과 같은 계산 과정을 반복하기로 합니다.  
 
    n → n / 2 (n이 짝수일 때)  
    n → 3 n + 1 (n이 홀수일 때)  
  
13에 대하여 위의 규칙을 적용해보면 아래처럼 10번의 과정을 통해 1이 됩니다.  
  
    13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1  
아직 증명은 되지 않았지만, 이런 과정을 거치면 어떤 수로 시작해도 마지막에는 1로 끝나리라 생각됩니다.   
(역주: 이것은 콜라츠 추측 Collatz Conjecture이라고 하며, 이런 수들을 우박수 hailstone sequence라 부르기도 합니다)  
  
그러면, 백만(1,000,000) 이하의 수로 시작했을 때 1까지 도달하는데 가장 긴 과정을 거치는 숫자는 얼마입니까?  
  
참고: 계산 과정 도중에는 숫자가 백만을 넘어가도 괜찮습니다.  

```python
def collatz_count(n):
    c = 1
    while n > 1:
        c += 1
        if n % 2 == 0:
            n = n / 2
        else:
            n = 3*n+1
    return c

max_num = 0
max_count = 0
count = 0
for i in range(1000001):
    count = collatz_count(i)
    if count > max_count:
        max_count = count
        max_num = i
print max_num
```

## 설명 

> 이 문제는 백만 이하의 수들이 1까지 도달하는 데 가장 긴 과정을 거치는 숫자를 찾는 것입니다. 일단 0부터 백만까지 해야하기 때문에 range를 1000001까지 범위를 지정합니다. 그리고 collatz_count라는 함수를 정의하는데, 이 함수는 인자로 들어온 수가 짝수이면, n = n/2를 하고 count를 하나 올리고, 홀수이면, n = 3 * n + 1을 하고 count를 하나 올립니다. 그 반복문은 n이 1보다 작을때까지 계속하게 됩니다. 그렇게 나온 count 값(c)가 반환되고, 그 값을 이전에 가장 큰 max_count 값과 비교하여 크기를 비교하고 더 크면, max_count에 그 값을 넣고, max_num 에는 i의 값을 넣고 다시 반복합니다. 그리고 반복문이 끝나면 max_num를 출력합니다.

프로젝트 오일러 문제 15  

아래와 같은 2 × 2 격자의 왼쪽 위 모서리에서 출발하여 오른쪽 아래 모서리까지 도달하는 길은 모두 6가지가 있습니다 (거슬러 가지는 않기로 합니다).


그러면 20 × 20 격자에는 모두 몇 개의 경로가 있습니까?  


```python

def path_count(x, y):
    count = 0
    n = 0
    if x == 0 or y == 0:
        return 1
    elif x == 1 or y == 1:
        return x + y
    else:
        if x > y:
            max_n = x
            min_n = y
        else :
            max_n = y
            min_n = x
            
        for i in range(min_n + 1):
            n = 0
            if i == 0 or i == max_n:
                count += 1
            else:
                n = (path_count(i, max_n - i) * path_count(i, min_n - i))
                count += n
            #print "%d번째: %d, %d" % (i, n, count)
        return count    
print path_count(20,20)
```

## 설명

> 이번 문제는 격자의 모서리의 경로 수를 구하는 것입니다. 이번 문제도 어렵지는 않았으나 조금 생각을 해야하는 문제였습니다. 일단, 인자 x, y를 받는 path_count 함수를 정의합니다. 이함수는 재귀함수로, 경로의 개수를 구하기 위한 함수입니다. 일단, x or y가 0일때는 1을 반환하고, x or y가 1일때는 x+y의 값을 반환합니다. 그리고 이 2경우가 아닌 모든 경우, x와 y의 크기에 따라 max_n, min_n에 각각의 값을 저장합니다. 그리고, 반복문을 통해 range(min_n+1)를 사용하여 i값을 반복합니다. 그리고 범위를 min_n(x,y값중 작은 값)으로 잡은 이유는 이후, 크기가 동일하지 않은 격자(예. 5 x 2)를 구할 때, min_n으로 기준을 잡고 구하기 때문입니다. 그 다음은 조건문을 통해서, i의 값이 0 or max_n 값일 경우에는 count 값을 1 올리게 합니다. (이는 가장 우측 또는 좌측 경로를 나타냅니다.) 그리고 해당하지않는 경우에는 count 값을 path_count(i, max_n - i) * path_count(i, min_n - i) 로 올리게됩니다. 이때 재귀 함수가 사용되며, 이 방법은 내부에 경로를 다시 작은 격자의 경로로 잡고 경로 개수를 얻는 방식입니다. 그리고 마지막으로 count 값을 반환하여 값을 출력합니다.

> 조건 4)는 조건 2)와 3)을 다를 사이트로 실행해야합니다. 그래서 저는 저번에 했던 코드를 다시 수정해서 한번에 값들이 나올 수 있게 정리해서 5개의 외국 사이트를 임의로 가지고 와서 파일로 저장하게 만들었습니다.

> 조건 5)에서는 words.frequency 파일에 들어있는 값들을 다시 객체로 가지고 온 후, 출현 빈도가 높은 단어 3개씩을 뽑아서 리스트로 저장했습니다. 리스트의 형태는 내부에 튜플로 저장하는데, 앞에는 사이트명,그리고 뒤는 사전 형태의 단어와 빈도수가 저장되어 있습니다. 

> 조건 6)에서는 사용자가 임의로 입력한 단어를 제가 임의의 방식으로 유사도를 결정한 후, 유사도의 크기에 따라 site를 나열하게 되어있습니다.
